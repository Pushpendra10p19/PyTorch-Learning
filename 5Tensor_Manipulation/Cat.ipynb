```python
# Cat()  -->  it is used to join the existing dimensions
import torch
a = torch.tensor([1,2,3])
b = torch.tensor([4,5,6])
torch.cat((a,b))

# 2D....
a = torch.tensor([
    [1,2,3],
    [4,5,6]
])
b = torch.tensor([
    [2,3,4],
    [5,6,7]
])
torch.cat((a,b)) # dim = 0 (default) for row wise join
torch.cat((a,b), dim = 1) # dim = 1 for column wise join

# if the dimansions are not same
#a = torch.tensor([
#    [1,2,3],
#    [4,5,6]
#])
#b = torch.tensor([
#    [2,3,4],
#    [5,6,7]
#])
#torch.cat((a,b)) # dim = 0 (default) for row wise join
#torch.cat((a,b), dim = 1) # dim = 1 for column wise join'''
```
