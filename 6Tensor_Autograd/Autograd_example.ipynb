```python
# Autograd -->  it is used in deep learning to give the right prediction by Automatic gradient calculation
# In PyTorch, `requires_grad` specifies whether or not the gradient needs to be calculated for a tensor

# Normal.....
import torch
a = torch.tensor(2.0)
print(a.requires_grad)

# Gradient tracking ON.....
x = torch.tensor(2.0, requires_grad=True)
print(x.requires_grad)

# Example-1.....
a = torch.tensor(3.0, requires_grad=True)
b = a * 2
c = b + 5
print(c)
```
