```python
# Backward()  -->  The 'backward' operation is used to calculate gradients by moving backwards.

#Example-1.........
import torch
x = torch.tensor(4.0, requires_grad=True)
y = x * 3
z = y ** 2
z.backward()
print(x.grad)

#Example-2......
a = torch.tensor(2.0, requires_grad=True)
b = 4 * a ** 2 + 3 * a + 1
b.backward()
print(a.grad)
```
