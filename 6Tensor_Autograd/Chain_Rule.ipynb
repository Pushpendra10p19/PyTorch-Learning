```python
# Chain Rule ---> Chain Rule is used to calculate the millions of weights and thousands of operations 
# using Autograd 

# [dz/dx = dz/dy * dy/dx]  it is the formula for human calculation because without chain rule human
# can not calculate of millions of weights and thousands of operation 

# Human can perform the normal Operations which is used in Autograd and Backward classes

import torch
x = torch.tensor(2.0, requires_grad=True)
y = x * 3
z = y ** 2
z.backward()
print(x.grad)
print(z.requires_grad)



```
