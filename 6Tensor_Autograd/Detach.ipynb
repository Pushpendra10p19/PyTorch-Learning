```python
# detach()  --> it is used to detached the connection from computational graph
# This means that gradients for operations performed on that tensor will not be tracked.

# Example...
import torch
x = torch.tensor(5.0, requires_grad=True)
y = x * 2
z = y.detach()
print(z.requires_grad)

# detach() does not change the value
print(y)
print(z)
```
