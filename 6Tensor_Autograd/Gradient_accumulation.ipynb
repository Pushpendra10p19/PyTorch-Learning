```python
# Gradient accumulation ---> gradient accumulation is the technique in which if we run again the backward()
# function then first Output and Second Output will add

# Example-1.....
import torch
x = torch.tensor(3.0, requires_grad=True)

# First backward()...
y = x**2
y.backward()
print(x.grad)

# Second backward()...
y = x**2

y.backward()

print(x.grad)

# Output = First + Second
```
